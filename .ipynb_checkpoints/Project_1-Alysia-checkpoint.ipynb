```python
#Below is the alternate way to pull census income data from the ACS5 using the URL
```

```python
years=['2010','2009','2008','2007','2006']
all_data=[]
for year in years:
    try:
        target_url = f"https://api.census.gov/data/{year}/acs/acs1?get=NAME,B01003_001E,B19013_001E&for=state:*&key={api_key}"
        income_data = requests.get(target_url).json()
        all_data.append({year:income_data})
    except:
        print('not found')
```
